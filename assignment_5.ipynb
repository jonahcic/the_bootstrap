**Q1.** A die is fair if every face is equally likely. A die has six sides if it has six faces labelled 1, 2, ... , 6.

1. Imagine rolling two dice, $d_1$ and $d_2$. Let $R_{min}$ be the lesser value of the two face values. What is the probability of getting a 1, 2, 3, 4, 5, or 6? Which values are more or less likely compared to the roll of a single six-sided die? What is the expected value of $R_{min}$? Plot the probability and cumulative distribution functions for $R_{min}$. Compute this by hand and simulate it using the law of large numbers.
2. Imagine rolling three dice, $d_1$, $d_2$, and $d_3$. Let $R_{med}$ be the middle of the three face values. So if you roll 2, 3 and 4, the middle value is 3, and if you roll 2, 4, 4, the middle value is 4, and so on. What is the probability of getting a 1, 2, 3, 4, 5, or 6? Which values are more or less likely compared to the roll of a single six-sided die? What is the expected value of $R_{med}$? Plot the probability and cumulative distribution functions for $R_{med}$. I recommend using simulations and the law of large numbers.
3. Imagine rolling a die. If you roll 1, 2, 3, 4, or 5, add that number to your total and stop; if you roll a six, add that number to your total and roll the die again. So you could roll, say, two sixes and then a four, and get a total of 16, or one three and get a total of 3, or twelve sixes and 1 and get 72, and so on. Write code to simulate this process, and determine its expected value using the law of large numbers. What is the probability of getting a total of 1, 2, 3, ... and so on, in your simulation? I recommend using simulations and the law of large numbers.  (Hint: The `while` loop might be useful in this case.)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
### 1

values = np.array([1,2,3,4,5,6])

# computing by hand

pdf = np.zeros(6)
for i in range(len(pdf)):
    for j in range(len(pdf)):
        min_ij = np.min([values[i],values[j]])
        pdf[min_ij-1] += 1
pdf = pdf/36
cdf = np.cumsum(pdf)

expected_value = np.sum(values*pdf)
print("Expected value: ", expected_value)

# bootstrap simulation w/ law of large numbers

num = 600
d1 = np.random.choice(values, size=num)
d2 = np.random.choice(values, size=num)
min_roll = [ min([d1[i], d2[i]]) for i in range(num) ]

ecdf = np.zeros(6)
for i in range(6):
    ecdf[i] = (1/num)*np.sum( min_roll <= values[i] )

# plotting PDF
plt.scatter(values,pdf)
plt.plot(values,pdf)
plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("Probability Density Function (PDF)")
plt.show()

# plotting  ECDF vs. CDF
plt.scatter(values,ecdf,label="ECDF")
plt.plot(values,ecdf,label="ECDF")
plt.scatter(values,cdf,label="CDF")
plt.plot(values,cdf,label="CDF")
plt.xlabel("X")
plt.ylabel("Probability")
plt.title("Empirical and Theoretical CDF")
plt.legend(loc="lower right")
plt.show()
```

Compared to the roll of a single six-sided die, in our scenario lower numbers (1, 2 and 3) are more likely to be R_min and higher numbers (4, 5 and 6) are less likely to be R_min. The expected value of R_min is 2.53. The probability of getting each respective number is as follows: 1=0.3056, 2=0.25, 3=0.1944, 4=0.1389, 5=0.0833, and 6=0.0278.

```python
### 2

# calculating by hand

pdf = np.zeros(6)
for i in range(len(pdf)):
    for j in range(len(pdf)):
        for k in range(len(pdf)):
          med_ijk = int(np.median([values[i],values[j],values[k]]))
          pdf[med_ijk-1] += 1
pdf = pdf/216
cdf = np.cumsum(pdf)

# simulating using bootstrap and law of large numbers

values = np.array([1,2,3,4,5,6])
num = 3000
d1 = np.random.choice(values, size=num)
d2 = np.random.choice(values, size=num)
d3 = np.random.choice(values, size=num)
med_roll = [int(np.median([d1[i], d2[i], d3[i]])) for i in range(num)]

expected_value = np.sum(med_roll)/num
print("Expected value: ", expected_value)

ecdf = np.zeros(6)
single = np.zeros(6)
for i in range(6):
    ecdf[i] = (1/num)*np.sum( med_roll <= values[i] )
    single[i] = (1/num)*np.sum( d1 <= values[i] )

# plotting pdf

plt.scatter(values,pdf)
plt.plot(values,pdf)
plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("Probability Density Function (PDF)")
plt.show()

# plotting CDF vs. ECDF

plt.scatter(values,cdf,label="CDF")
plt.plot(values,cdf,label="CDF")
plt.scatter(values,ecdf,label="ECDF")
plt.plot(values,ecdf,label="ECDF")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("Empirical and Theoretical CDF")
plt.legend(loc="lower right")
plt.show()

# plotting single vs. median roll

plt.scatter(values,ecdf,label="Median")
plt.plot(values,ecdf,label="Median")
plt.scatter(values,single,label="Single")
plt.plot(values,single,label="Single")
plt.xlabel("X")
plt.ylabel("Probability")
plt.legend(loc='lower right')
plt.title('Median and Single Roll CDF')
plt.show()
```

Compared to the roll of a single six-sided die, in our scenario the middle numbers (3 and 4) are more likely to be R_med.

```python
### 3

values = np.array([1,2,3,4,5,6])

num = 600
results = []

for i in range(num):
    roll_again = True
    sum = 0
    while roll_again is True:
        roll = np.random.choice(values, size=1)
        sum += roll
        if roll < 6:
            roll_again = False
            results.append(int(sum[0]))

results = pd.DataFrame(results)

expected_value = np.mean(results)
print("Expected value: ", expected_value)

# plotting histogram of simulation results
sns.histplot(results)
plt.show()

# plotting ECDF
sns.ecdfplot(results)
plt.title("Empirical CDF")
plt.show()
```

```python
### 1

df = pd.read_csv("/content/the_bootstrap/data/mammogram.csv")

print(pd.crosstab( df["treatment"], df["breast_cancer_death"]) , "\n")
print(pd.crosstab( df["treatment"], df["breast_cancer_death"], margins=True, normalize=True) , '\n')

diff = 44425/44925 - 44405/44910
print("Difference Controlvs. Mammogram: ", diff)
```

```python
### 2

df["survive"] = 0
df.loc[ df["breast_cancer_death"] == "no", "survive"] = 1

df_treat = df.loc[ df["treatment"] == "mammogram", :]
df_control = df.loc[ df["treatment"] == "control", :]

num = 5000
effect = []
for i in range(num):
    df_treat_i = df_treat.sample( df_treat.shape[0], axis=0, replace=True)
    df_control_i = df_control.sample( df_control.shape[0], axis=0, replace=True)
    treatment_effect_i = np.mean(df_treat_i["survive"]) - np.mean(df_control_i["survive"])
    effect.append(treatment_effect_i)

sns.kdeplot(effect)
plt.show()

sns.ecdfplot(effect)
plt.show()
```

```python
### 3

level = .99
lower_bound = np.quantile(effect, 1-level)
upper_bound = np.quantile(effect, level)

print('99% confidence interval: (', lower_bound, ', ', upper_bound, ')' )
```

Yes, the 99% confidence interval includes zero.
